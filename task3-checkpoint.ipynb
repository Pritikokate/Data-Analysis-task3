```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('C:\\Users\\kokat\\Downloads\\cust\\Mall_Customers.csv')
```

```python
# Check basic info
print(df.info())

# Check shape
print("Dataset Shape:", df.shape)

# Check missing values
print("Missing Values:\n", df.isnull().sum())

# Check for duplicates
print("Duplicates:", df.duplicated().sum())

# Summary statistics
print(df.describe())
```

```python
features = df[['Age', 'Annual_Income_(k$)', 'Spending_Score']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_clusters = 5  # Replace with the selected number
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)
df['Cluster'] = cluster_labels
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title('Customer Segments')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df[['Age', 'Annual_Income_(k$)', 'Spending_Score', 'Cluster']], hue='Cluster', palette='viridis')
plt.show()
```

```python
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], color='red', marker='X', s=200)
plt.title('Clusters with Centroids')
plt.show()
```
